# KeyBert annotation
In this section, we will extract keywords from the 100 vacancies using KeyBERT. This stage serves three purposes:
1. To fine-tune the KeyBERT module in subsequent stages.
2. To compare supervised keyword extraction with automated keyword extraction.
3. To demonstrate improvement due to fine-tuning.

```python
df_vacancies_100["KeyBert"]=""
```

```python
# Initialize the KeyBERT model
kw_model = KeyBERT()

# Define a function to extract keywords
def extract_keywords(description):
    keywords = kw_model.extract_keywords(description, keyphrase_ngram_range=(1, 2), stop_words='english')
    # Extract just the keywords (not the scores) and join them as a set
    keywords_set = set([kw[0] for kw in keywords])
    return keywords_set

# Apply the function to each row in the 'job_description' column and store results in 'KeyBert' column
df_vacancies_100['KeyBert'] = df_vacancies_100['job_description'].apply(extract_keywords)

# Display the updated DataFrame
df_vacancies_100.head()
```

# Anotated data
The following imported data set is anotated and include the following:


1.   job_num: Numerical identifier for job descriptions.
2.   job_description: Contains detailed descriptions of job roles
3.   word: Lists individual words or phrases extracted from the job descriptions.
4.   hadar: annotation column with binary values (0 or 1), this column meant to label the words as keyword(1) or non-keyword(0).
5.   nisan: annotation column with binary values (0 or 1), this column meant to label the words as keyword(1) or non-keyword(0).
6.   judge: 3rd annotator which steps in only when there is a disagreement 1st annotator to the 2nd annotator.
7.   final: labels of the words. combined from judge column and agreemnt of 1st and 2nd anotator.
8.   chatgpt: for each new position the chatgpt extract the keywords from the job description. later there are binary values to determine which word he indetifed as keyword as expected. 1=success 0=fail.
9.    KeyBert: for each new position the KeyBert model extract the keywords from the job description. later there are binary values to determine which word he indetifed as keyword as expected. 1=success 0=fail. *please be advised: the KeyBert model hasnt been tuned yet.

```python
import sklearn.metrics
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
x=data['hadar']
y=data['nisan']
hadar_filtered = x[x!='-']
nisan_filtered = y[y!='-']
hadar1=hadar_filtered.to_list()
nisan1=nisan_filtered.to_list()
if len(hadar1)==len(nisan1):
  print('yes')
x=hadar1
y=nisan1
```

```python
print(f"length of annotation column: {len(data)-100}") # Subtraction -100 for 100 vacencies
count_1_x = x.count(1)
count_0_x = x.count(0)
count_1_y = y.count(1)
count_0_y = y.count(0)
# Count the number of agreements
agreements = sum(1 for a, b in zip(x, y) if a == b)
print(f"Number of agreements between 'hadar' and 'nisan': {agreements}")
# Print the results
print(f"Count of 1s in 'hadar': {count_1_x}")
print(f"Count of 0s in 'hadar': {count_0_x}")
print(f"Count of 1s in 'nisan': {count_1_y}")
print(f"Count of 0s in 'nisan': {count_0_y}")
kappa_score = cohen_kappa_score(x, y) #x is hadar y is nisan
print("Cohen's kappa score:", kappa_score)
```

```python
a = data['final']
b = data['KeyBert']
print(len(a), " ", len(b))

label_filtered = a[a != '-']
keybert_filtered = b[(b == 0) | (b == 1)]

label1 = label_filtered.to_list()
keybert1 = keybert_filtered.to_list()

if len(label1) == len(keybert1):
    print('yes')

a = label1
b = keybert1

print(len(a), " ", len(b))
```

```python
print("keybert accuracy score: ",accuracy_score(a, b))
print("keybert f1 macro score: ",f1_score(a, b, average='macro'))
```

```python
a = data['final']
c = data['chatgpt']
print(len(a), " ", len(a))

label_filtered = a[a != '-']
chatgpt_filtered = c[(c == 0) | (c == 1)]

label1 = label_filtered.to_list()
chatgpt1 = chatgpt_filtered.to_list()

if len(label1) == len(chatgpt1):
    print('yes')

a = label1
c = chatgpt1

print(len(a), " ", len(c))
```

```python
print("chatgpt accuracy score: ",accuracy_score(a, c))
print("chatgpt f1 macro score: ",f1_score(a, c, average='macro'))
```
